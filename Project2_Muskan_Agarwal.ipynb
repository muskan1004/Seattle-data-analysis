```python
df = df_2015[['ENERGYSTARScore',
       'SiteEUIWN(kBtu/sf)', 'SourceEUIWN(kBtu/sf)', 'SiteEnergyUseWN(kBtu)', 'SteamUse(kBtu)', 'Electricity(kBtu)',
       'NaturalGas(kBtu)', 'OtherFuelUse(kBtu)','GHGEmissions(MetricTonsCO2e)','GHGEmissionsIntensity(kgCO2e/ft2)']]
```

```python
df.info()
```

```python
df['ENERGYSTARScore'].describe()
```

Filling the missing values of Energy Star Scores with the mean values

```python
mean=67.894141
df['ENERGYSTARScore'].fillna(value=mean,inplace=True)
```

```python
df.dropna(inplace=True)
```

```python
df.head()
```

SPLITTING AND TRAINING THE DATA

```python
X=df[['ENERGYSTARScore', 'SiteEUIWN(kBtu/sf)', 'SourceEUIWN(kBtu/sf)',
       'SiteEnergyUseWN(kBtu)', 'SteamUse(kBtu)', 'Electricity(kBtu)',
       'NaturalGas(kBtu)', 'OtherFuelUse(kBtu)',
       'GHGEmissions(MetricTonsCO2e)']]
```

```python
y=df['GHGEmissionsIntensity(kgCO2e/ft2)']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
X_train, X_test, y_train, y_test = train_test_split( X, y, test_size=0.30, random_state=42)
```

```python
lm=LinearRegression()
```

```python
lm.fit(X_train,y_train)
```

```python
print(lm.intercept_)
```

PREDICTIONS

```python
predictions = lm.predict(X_test)
```

```python
plt.scatter(y_test,predictions)
plt.xlabel('Y Test')
plt.ylabel('Predicted Y')
```

```python
import numpy as np
```

```python
from sklearn import metrics

print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, predictions))
print('Mean Squared Error:', metrics.mean_squared_error(y_test, predictions))
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, predictions)))
```
